```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url='https://wuzzuf.net/search/jobs/?a=navbl%7Cspbl&q=data%20scientist&start=0'

response=requests.get(url)
soup = BeautifulSoup(response.content, 'lxml')
job_titles=soup.find_all('h2',{'class':'css-m604qf'})
job_titles=[job.text for job in job_titles]
company_names=soup.find_all('a',{'class':'css-17s97q8'})
company_names=[company.text for company in company_names]
locations=soup.find_all('span',{'class':'css-5wys0k'})
locations=[location.text for location in locations]
job_types=soup.find_all('div',{'class':'css-1lh32fc'})
job_types=[job.text for job in job_types]
links=soup.find_all('a',{'class':'css-o171kl'})
links=['https://wuzzuf.net'+link['href'] for link in links]
df=pd.DataFrame({'job_titles':job_titles,'company_names':company_names,'locations':locations,'job_types':job_types})
df.to_csv('jobs.csv')
df
```
